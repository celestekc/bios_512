````r
p = ggplot (penguins_long, aes(x=species, y=value, color=sex))
p = p + geom_point(position = position_jitter(h = 0, w = 0.20), size = 1)
p = p + theme(axis.text.x = element_text(angle = 45, hjust = 1))
p = p + facet_wrap(~name, nrow = 1, scales = "free")
p
````

````r
p = p + stat_summary(fun = median, 
                     fun.min = median, 
                     fun.max = median, 
                     geom = "crossbar", 
                     width = 0.75, 
                     size = 0.2,
                     mapping = aes(group = species))

p = p + theme(axis.text.x = element_text(hjust = 1, angle = 45))
p
````
